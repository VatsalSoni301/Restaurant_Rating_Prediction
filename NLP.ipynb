```python
df = pd.read_csv("yelp.csv")

texts = df['text']
texts = texts.str.lower()
```

```python
from nltk.stem.snowball import SnowballStemmer
stemmer = SnowballStemmer("english")
stemmer
```

```python
def stemWords(tokenizedWords):
    for i in range(1, len(tokenizedWords) + 1):
        # filteredWords=[word for word in tokenizedWords[i] if word not in stopwords.words('english')]
        for word in tokenizedWords[i]:
#             print(filteredWords)
            filteredWords = [word for word in tokenizedWords[i] if word in stemmer.stem(word)]
            tokenizedWords[i] = filteredWords
    return tokenizedWords
```

```python
from nltk.corpus import stopwords
stop = stopwords.words('english')
# print(len(stop))
stop=stop[0:144]
# print(stop)
```

```python
# Exclude stopwords with Python's list comprehension and pandas.DataFrame.apply.
texts = texts.apply(lambda x: ' '.join([word for word in x.split() if word not in (stop)]))
# df['text']=
```

```python
texts = texts.apply(lambda x:' '.join([stemmer.stem(word) for word in x.split() if word in stemmer.stem(word)]))
```

```python
# texts = stemWords(texts)
```

```python
positive = 0
negative = 0
```

```python
for i in texts:
    blob = TextBlob(i)
    feel = blob.sentiment
    if feel.polarity > 0:
        positive = positive + 1
    else:
        negative = negative + 1
```
